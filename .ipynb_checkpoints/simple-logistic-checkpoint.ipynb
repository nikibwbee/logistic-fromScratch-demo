```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_csv('simple_logistic.csv')
data.head()
```

```python
plt.scatter(data.Age, data.Bought, marker='+', color='red')
plt.show()
plt.xlabel("bought")
plt.ylabel("age")
```

```python
x = data['Age'].tolist()
y= data['Bought'].tolist()
```

```python
#split--
import random

dataset= list(zip(x,y))
random.shuffle(dataset)

train_size = int(0.75 * len(dataset))
train_data = dataset[: train_size]
test_data=dataset[train_size:]

x_train = [i[0] for i in train_data]
y_train = [i[1] for i in train_data]
x_test = [i[0] for i in test_data]
y_test = [i[1] for i in test_data]
```

```python
import math
def sigmoid(z):
    return 1/ (1+math.exp(-z))
```

```python
w=0.0
b=0.0

learning_rate = 0.01
epochs = 20000
m=len(x_train)


```

```python
#gradient descent

for epoch in range(epochs):
    dw=0.0
    db=0.0

    for i in range(m):
        z= w* x_train[i] +b
        pred =sigmoid(z)
        error = pred - y_train[i]
        dw += error *x_train[i]
        db+= error
        
    dw /= m
    db /=m

    w -=learning_rate * dw
    b -= learning_rate * db

print("trained weight: ", w)
print("trained bais: " , b)
```

```python
def predict(age):
    prob = sigmoid(w*age+b)
    return 1 if prob >=0.5 else 0
    
```

```python
# test predictions.
for age, label in zip(x_test,y_test):
    pred = predict(age)
    prob =sigmoid(w*age +b)
    print(f"Age {age}: Actual {label}, Probability {prob:2f},Predicted {pred}")
    
```
